```python
import xarray as xr
import glob

input_folder = '/g/data/u46/users/ext547/ewater/input_data/temperature/temperature_nc/' #location of files
ncfiles = glob.glob(input_folder + 'daily/*.nc') #a glob file containing the names of all nc
```

```python
for ncpath in ncfiles:
    nc_name = ncpath[-38:]
    print(nc_name)
    print(ncpath)
    ds = xr.open_dataset(ncpath, drop_variables = 'timestp')#,decode_times=False)
    ds.time.attrs['long_name'] = 'Time, unix time-stamp'
    ds.to_netcdf(input_folder+'daily_mean/'+nc_name, #mode='w',
              encoding={'time': {
                  'units': "seconds since 1970-01-01",
                  'calendar': "standard",
                  'dtype': "float64"
              }})
```
