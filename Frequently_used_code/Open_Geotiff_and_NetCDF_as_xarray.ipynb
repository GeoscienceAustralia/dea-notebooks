# Opening a Geotiff with Xarray

**Notebook currently compatible with Both `NCI` and `DEA Sandbox` environment**

### Background
It may often be useful open an external dataset, or data that you have previously saved to NetCDF or geotiff, into a notebook in order to conduct analysis on it or combine it with DEA data. In this example we will open a section of the NVIS major vegitation groups dataset published by DoEE avalible (https://www.environment.gov.au/land/native-vegetation/national-vegetation-information-system). We will also open some Fractional Cover data which was previously saved to NetCDF.

for more information on the libraries used:
* xarray.open_rasterio (http://xarray.pydata.org/en/stable/generated/xarray.open_rasterio.html)
* xarray.open_dataset (http://xarray.pydata.org/en/stable/generated/xarray.open_dataset.html)

### Description
This notebook shows how to open raster data from file into Xarray using Xarray's built in fuctions for handeling these data types:
* opeing a geotiff into xarray 
* opeining a NetCDF file into xarray



## Getting started
 To run this example, run all the cells in the notebook, starting with the "Load packages" cell. 

### Load packages


```python
%matplotlib inline

import datacube
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

### Connect to the datacube


```python
dc = datacube.Datacube(app="Open_geotiffs_and_NetCDF_in_xarray")
```

## Define path files


in the code below we define the locations of the Geotiff NetCDF files that we will open

```python
# Use code comments for low-level documentation of code
Geotiff_location = '.../Datasets/NVIS_files/NVIS_major_veg_groups_ACT.tif'

NetCDF_location = '.../Datasets/9_-20/LS_FC_MED_3577_9_-20_20080101_20090101.nc'
```

## Open The Geotiff


To open a geotiff we use Xarray.open_rasterio(). this fuction is built around rasterio and can also be used to open large geotiffs as a DaskArray by providing a **chunks**=(int,tuple or dict) parameter. 

```python
#open into an xarray DataArray
data1 = xr.open_rasterio(Geotiff_location)

data1
```

```python
# Plot the data with Xarray built in plot fuction
data1.plot(figsize =(13,15), cmap='gist_ncar_r', vmax=70)
```

## Open the NetCDF

To open a NetCDF we use Xarray.open_dataset(). This can also be used to open large geotiffs as a DaskArray by providing a **chunks**=(int or dict) parameter. 

```python
#open into an xarray DataArray
data2 = xr.open_dataset(NetCDF_location)

data2
```

```python
# Plot the data with Xarray built in plot fuction
data2.BS.plot(figsize =(12,10))
```
