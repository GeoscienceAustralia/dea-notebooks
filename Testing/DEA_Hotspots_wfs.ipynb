```python
import geopandas as gpd

# Specify bounding box
ymax, xmin = -34.240120, 147.974369
ymin, xmax = -35.326919, 149.107405

hotspots_wfs = f'https://hotspots.dea.ga.gov.au/geoserver/wfs?' \
                       f'service=WFS&version=1.1.0&request=GetFeature' \
                       f'&typeName=hotspots&maxFeatures=100000' \
                       f'&bbox={ymin},{xmin},{ymax},{xmax},' \
                       f'urn:ogc:def:crs:EPSG:4326'

hotspots_gdf = gpd.read_file(hotspots_wfs)
hotspots_gdf.start_dt.dt.year.plot.hist()
```

```python
hotspots_gdf.plot(column=hotspots_gdf.start_dt.dt.year)
```
